Sea f(x)= x^2 -5, encontar la raiz que se encuentra en el intervalo [2,3] con un error permisible porcentual de 0.1%

```
import matplotlib.pyplot as plt
```

```
def f(x):
  return (x**2)-5
```

```
def Compute_xr(xl,xu):
  return (xl+xu)/2
```

```
def Biseccion (xl,xu,criterioparo=0.1,Maxiter=100):
  status= False
  xl= xl
  xu= xu
  count= 0
  e_perm= criterioparo
  error= xu
  maxiter= Maxiter 
  xr_vec=[]
  xu_vec=[]
  xl_vec=[]
  err_vec=[]

  while (count <= maxiter):
    xr= Compute_xr (xl ,xu) 
    xr_vec=[] .append(xr)
    xu_vec=[] .append (xu)
    xl_vec=[] .append (xl)

    f_xl=f(xl)
    f_xr= f(xr)
    f_xu=f(xu)

    err_vec. append(abs((xu-xl)/(xu+xl)))

    if (f(xl)*f(xr)>0):
      xl=xr
    if (f(xu)*f(xr)>0):
      xu=xr

    if abs(err_vec[count] - error) < 0.0000001:
      status= True
      break
    
    count+=1
    error= err_vec[count-1]

  if status:
    plt.plot(xl_vec,label= '$x_l$')
    plt.plot(xu_vec,label= '$x_u$')
    plt.plot(xr_vec,label= '$x_r$')
    plt.legend()
    plt.show()

  return [xr_vec[count-1],status]
```

```
xl = 2 
xu = 3
[xr, status]=Biseccion(xl, xu)

if status: 
   print("La raiz (xr) es: ", xr)
else:
   print("ERROR")
```
